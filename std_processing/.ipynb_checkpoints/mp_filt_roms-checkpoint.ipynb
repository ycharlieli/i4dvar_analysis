```python
import xarray as xr
import numpy as np
import scipy as sp
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import os
from dask.distributed import Client
import seapy
import sys
import time
from lanczos_filt import *
import warnings
warnings.filterwarnings('ignore')
import os
import gc
# os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
sys.setrecursionlimit(100000)
```

```python
Tcount = {} 
Var_std = {}
Var_mean = {}
x_chunk = int(262/2)
y_chunk = int(362/2)
z_chunk = 10
var_list = ['ubar','vbar']
for i in range(1,13):
    Tcount[i] = {}
    Var_std[i] = {}
    Var_mean[i] = {}
    for iv in var_list:
        Tcount[i][iv] = 0

mon_indata = []
Find = 172
Fstep = 5 # file number depends on the machine IO ability. Here 5 is optimal , beyond this  IO ability will contraint CPU efficiency
for i in range(55):
    Inp_files = ['/Volumes/TO_1/outputs_velbar/ocean_ecs_avg_'\
                   +str(i).rjust(4,'0')+'_t.nc' for i in range(Find,Find+Fstep)]
    template_file = '/Volumes/TO_1/outputs_SCORRECTION/ocean_ecs_avg_0001.nc'
#        Inp_files = ['F:\\outputs_SCORRECTION\\ocean_ecs_avg_'\
#                      +str(i).rjust(4,'0')+'_t.nc' for i in range(Find,Find+Fstep)]
#        template_file = 'E:\\outputs_SCORRECTION\\ocean_ecs_avg_0001.nc'
    #Inp_files = ['E:\\outputs_SCORRECTION\\ocean_ecs_avg_'\
    #             +str(i).rjust(4,'0')+'.nc' for i in range(180,447)]
    #print(len(Inp_files))
    print('loading ')
    [print(ifile) for ifile in Inp_files]
    fields = set(seapy.roms.fields)

    grid = seapy.model.asgrid(template_file)
    nc = seapy.netcdf(template_file)

    fields = set(nc.variables).intersection(fields)

    epoch, time_var = seapy.roms.get_reftime(nc)

    

    Inp_ds = xr.open_mfdataset(Inp_files,
                                coords='minimal',compat='override',
                                # data_vars=['temp','salt','u','v','zeta'],
                               parallel=True,
                               chunks={'eta_rho':y_chunk,'xi_rho':x_chunk,'s_rho':z_chunk,
                               'eta_u':y_chunk,'xi_u':x_chunk,\
                               'eta_v':y_chunk,'xi_v':x_chunk,\
                               'eta_psi':y_chunk,'xi_psi':x_chunk,},
                               engine = 'netcdf4' ).chunk(dict(ocean_time=-1))

    # global Cf, Nf, M, Var_da, v
    dt = 1
    fs = 1/(dt)
    Nf = fs/2

    Cf = fs/36
    M = 100
    for v in var_list:
        # client.restart()

        print(v)
        client = Client(threads_per_worker=8, n_workers=4,memory_limit= '8 GiB')
        print(client)
        print(client.dashboard_link)
        Var_da = Inp_ds[v]
        # Var_da
        #        global Var_da
        # print(Var_da.shape)
        if v in ['zeta', 'ubar','vbar']:
            (ly,lx,lt) = Var_da.shape
            shape = Var_da.shape
            lz = 1
        else:
            (ly,lx,lz, lt) = Var_da.shape 
            shape = Var_da.shape

        
            
       #
        #for v in fields:
        #with futures.ThreadPoolExecutor(max_workers=8) as executor:
        #    for iy in range(ly):
        #        for ix in range(lx):
        #            executor.submit(mp_filt_var,Var_da,ix,iy,Cf,Nf,M)
                # p.map(mp_filt_var, zip(iter_x,iter_y))
    #    threads = []
    #    count = 0
    #    for iy in range(ly):
    #        for ix in range(lx):
    #            p.apply_async(mp_filt_var, args=(Var_da,Var_filt,ix,iy,Cf,Nf,M))
    #            t = p.apply_async(mp_filt_var,args =(Var_da,ix,iy,Cf,Nf,M))
        #        threads.append(t)
        #        count +=1
        print('Filtering...')
        start_total = time.time()
        start = time.time()
        Var_filt = xr.apply_ufunc(lanczos_filter,Var_da,Cf,Nf,M,
                                   input_core_dims=[["ocean_time"],[],[],[]],
                                   output_core_dims=[["ocean_time"]],
                                   vectorize=True,
                                   dask="parallelized",
                                   dask_gufunc_kwargs={'allow_rechunk':True},
                                   output_dtypes=[Var_da.dtype]
                                  ).data.compute()
        end = time.time()
        print('costing %f min'%((end-start)/60))
        client.close()
        if v == 'zeta':
            Var_filt = Var_filt[:,:,48:-47]
            Filt_da = xr.DataArray(Var_filt,
                                   coords = {'eta_rho':Inp_ds.eta_rho.data,
                                             'xi_rho': Inp_ds.xi_rho.data,
                                             'ocean_time': Inp_ds.ocean_time.data[48:-47]}
                                   ).chunk({'eta_rho':y_chunk,'xi_rho':x_chunk})
        elif v in ['temp', 'salt']:
            Var_filt = Var_filt[:,:,:,48:-47]
            Filt_da = xr.DataArray(Var_filt,
                                   coords = {'eta_rho':Inp_ds.eta_rho.data,
                                             'xi_rho': Inp_ds.xi_rho.data,
                                             's_rho': Inp_ds.s_rho.data,
                                             'ocean_time': Inp_ds.ocean_time.data[48:-47]}
                                   ).chunk({'eta_rho':y_chunk,'xi_rho':x_chunk,'s_rho':z_chunk})
        elif v == 'u':
            Var_filt = Var_filt[:,:,:,48:-47]
            Filt_da = xr.DataArray(Var_filt,
                                   coords = {'eta_u':Inp_ds.eta_u.data,
                                             'xi_u': Inp_ds.xi_u.data,
                                             's_rho': Inp_ds.s_rho.data,
                                             'ocean_time': Inp_ds.ocean_time.data[48:-47]}
                                   ).chunk({'eta_u':y_chunk,'eta_u':x_chunk,'s_rho':z_chunk})
        elif v == 'v':
            Var_filt = Var_filt[:,:,:,48:-47]
            Filt_da = xr.DataArray(Var_filt,
                                   coords = {'eta_v':Inp_ds.eta_v.data,
                                             'xi_v': Inp_ds.xi_v.data,
                                             's_rho': Inp_ds.s_rho.data,
                                             'ocean_time': Inp_ds.ocean_time.data[48:-47]}
                                   ).chunk({'eta_v':y_chunk,'eta_v':x_chunk,'s_rho':z_chunk})
        elif v == 'ubar':
            Var_filt = Var_filt[:,:,48:-47]
            Filt_da = xr.DataArray(Var_filt,
                                   coords = {'eta_u':Inp_ds.eta_u.data,
                                             'xi_u': Inp_ds.xi_u.data,
                                             'ocean_time': Inp_ds.ocean_time.data[48:-47]}
                                   ).chunk({'eta_u':y_chunk,'xi_u':x_chunk})    
        elif v == 'vbar':
            Var_filt = Var_filt[:,:,48:-47]
            Filt_da = xr.DataArray(Var_filt,
                                   coords = {'eta_v':Inp_ds.eta_v.data,
                                             'xi_v': Inp_ds.xi_v.data,
                                             'ocean_time': Inp_ds.ocean_time.data[48:-47]}
                                   ).chunk({'eta_v':y_chunk,'xi_v':x_chunk})           

        
        Filt_mongrp = Filt_da.groupby('ocean_time.month')
        
        for imonth in Filt_mongrp.groups.keys():
            print('month: ', imonth)
            print('accumulating data...')
            start = time.time()
            Filtvar_thismon = Filt_mongrp[imonth].data.compute(schedular='multiprocessing', num_workers=8).astype(np.float64) #Var_filt[:,:,:,Filt_mongrp.groups[imonth]] #  # #
            # Filtvar2_thismon = Filt_mongrp[imonth]**2 # (Var_filt[:,:,:,Filt_mongrp.groups[imonth]])**2
            # print(Filtvar_thismon)
            # print(Filtvar2_thismon)
            
            val_sum = Filtvar_thismon.sum(axis=-1)  # (Filt_mongrp[imonth].data.compute()).sum(axis=-1) #np.sum(Filtvar_thismon,axis=-1)
            val_sum2 = (Filtvar_thismon**2).sum(axis=-1) #np.sum(Filtvar2_thismon,axis=-1)
            # print(val_sum)
            # print(val_sum2)
            if Tcount[imonth][v] == 0:
#                    print('calculating std...')
#                    print(Filt_mongrp.groups[imonth])
                Var_std[imonth][v]  = val_sum2
#                        print('calculating mean...')
                Var_mean[imonth][v]  = val_sum
            else:
#                    print('calculating std...')
                Var_std[imonth][v] = Var_std[imonth][v] + val_sum2
#                   print('calculating mean...')
                Var_mean[imonth][v] = Var_mean[imonth][v] + val_sum
            end = time.time()
            print('costing %f min'%((end-start)/60))
            
            
            print('\tdata length of this month in this set: ', Filt_mongrp[imonth].ocean_time.shape[0])
            Tcount[imonth][v] +=  Filt_mongrp[imonth].ocean_time.shape[0]
            print(Filt_mongrp[imonth].ocean_time.shape[0])
            # sys.exit()
            del Filtvar_thismon
            # del Filtvar2_thismon
            del val_sum
            del val_sum2
            gc.collect()
        end_total = time.time()  
        print('total cost %f min'%((end_total - start_total)/60))
        del Var_da
        del Var_filt
        del Filt_da
        del Filt_mongrp
        gc.collect()
        
        # time.sleep(60)
    Find += Fstep

    
for imonth in Var_std.keys():
    print('month: ', imonth)
    for v in var_list:
        
        print(Tcount[imonth][v])
        if Tcount[imonth][v] is not 0 :
            Var_mean[imonth][v] = Var_mean[imonth][v]/Tcount[imonth][v]
            # plt.pcolor(Var_mean[imonth][v][:,:,-1])
            # plt.colorbar()
            # plt.show()
            Var_std[imonth][v] = np.sqrt(Var_std[imonth][v]/(Tcount[imonth][v]-1) - \
                                        Tcount[imonth][v]*(Var_mean[imonth][v]**2)/(Tcount[imonth][v] -1))
            # plt.pcolor(Var_std[imonth][v][:,:,-1])
            # plt.colorbar()
            # plt.show()
    sp.io.savemat('./bg_std/roms_ecs_v4sc_velbarstd_i_%d.mat'%(imonth),Var_std[imonth])
```
